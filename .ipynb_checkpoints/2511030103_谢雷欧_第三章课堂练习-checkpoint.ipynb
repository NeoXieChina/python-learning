```python
import math
a = float(input("请输入三角形的边长a："))

b = float(input("请输入三角形的边长b："))

c = float(input("请输入三角形的边长c："))

h = (a+b+c)/2

area = math.sqrt(h*(h-a)*(h-b)*(h-c))

print("三角形的面积为：{:.2f}".format(area))
```

# 3.问题：输入三角形三边，求三角形面积，对用户输入的数据做合法性检查。

```python
import math
a = float(input("请输入三角形的边长a："))
b = float(input("请输入三角形的边长b："))
c = float(input("请输入三角形的边长c："))

if a+b>c and a+c>b and b+c>a and a>0 and b>0 and c>0 and a-b<c and a-c<b and b-c<a:
    h = (a+b+c)/2
    area = math.sqrt(h*(h-a)*(h-b)*(h-c))
    print("三角形的面积为：{:.2f}".format(area))
else:
    print("您输入的数据有误，不是三角形！请重新输入！")
```

# 4.问题：根据用户的身高和体重，计算用户的BMI值，并给出相应的健康建议。
## BMI，即身体质量指数，是用体重（千克）除以身高（米）的平方得出的数字（BMI=体重(KG)÷身高2（M）），是目前国际上常用的衡量人体胖瘦程度以及是否健康的一个标准。我们先来看看成人的BMI数值：
### 成人的BMI数值
### 过轻：    低于18.5
### 正常：    18.5-23.9
### 过重：    24-27.9
### 肥胖：    28-32
### 过于肥胖：    32以上

```python
height = eval(input("请输入您的身高（米）："))
weight = eval(input("请输入您的体重（千克）："))
bml = weight/(height*height)
if bml<18.5:
    stand = "过轻"
elif bml<23.9:
    stand = "正常"
elif bml<27.9:
    stand = "过重"
elif bml<32:
    stand = "肥胖"
else:
    stand = "过于肥胖"
print("您的身高为"+str(height)+"米，您的体重为"+str(weight)+"千克，您的BMI为"+str(bml)+"，经过系统判断，您当前"+stand+"。")
```

# 1. 问题：统计英文句子中大写，小写和数字各有多少个？

```python
inputStr = str(input("请输入一串英文字符："))
countUpper=countLower=countNumber=0
for s in inputStr:
    if s.isupper():
        countUpper+=1
    elif s.islower():
        countLower+=1
    elif s.isdigit():
        countNumber+=1
print("在您输入的字符串",inputStr,"中，小写字符共",countLower,"个，大写字符共",countUpper,"个，数字字符共",countNumber,"个。")
```

# 2.问题：利用for循环求1至100中奇数和偶数的和分别是多少。

```python
sumOdd = sumEven = 0
startNumber = 1
endNumber = 100
for intNumber in range(startNumber,endNumber+1):
    if intNumber>0:
        if intNumber%2 == 0:
            sumEven+=intNumber
        else:
            sumOdd+=intNumber
print("在",startNumber,"-",endNumber,"中，奇数之和是",sumOdd,"，偶数之和是",sumEven,"。")
```

# 5.问题：求非负数字序列中的最小值、最大值和平均值。用户输入-1就表示序列终止。

```python
listNmuber = [0]
print("请输入非负整数，如果输入-1就停止输入！")
inputNumberList = eval(input("请输入一个数:"))
totalNumber = 0
sumNumber = 0
maxNumber = minNmuber = inputNumberList
while inputNumberList != -1:
    totalNumber += 1
    sumNumber += inputNumberList
    if inputNumberList > maxNumber:
        maxNumber = inputNumberList
    if inputNumberList < minNmuber:
        minNmuber = inputNumberList    
    inputNumberList = eval(input("请输入下一个数:"))
print("您输入的数的平均值是",sumNumber/totalNumber,"，其中最大值是",maxNumber,",最小值是",minNmuber,"。")
```

# 6.问题： “牛顿迭代法”求平方根。
## 1）假设需要求实数x的平方根，任取y为某个实数值
## 2）如果y×y=x，计算结束，y就是x的平方根
## 3）否则令z=(y+x/y)/2
## 4）令y的新值等于z，转回步骤1）
## y=(y+x/y)/2

```python
import math
x = float(input("请输入一个数赋值给x:"))
y = 1.0
while abs(y*y - x)>10e-8:
        y = (y+x/y)/2
print("根据牛顿迭代法计算，",x,"的平方根是",y,"。")
print("使用函数计算，",x,"的平方根是",math.sqrt(x),"。")
```

# break 和 continue 语句
# 7.问题：判断一个正整数n（n>=2）是否为素数。素数又称质数。一个大于1的自然数，除了1和它自身外，不能被其它整数整除的数叫做素数；否则称为合数。

### break版本

```python
n = int(input("请输入一个正整数："))
startNumber = 2
for i in range(startNumber,n):
    if n%i == 0:
        break
if i == n-1:
        print(n,"是素数。")
else:
        print(n,"是合数。")
```

### continue版本

```python
n = int(input("请输入一个正整数："))
startNumber = 2
isPrime = True
for i in range(startNumber,n):
    if n%i == 0:
        isPrime = False
        continue
if isPrime == True:
        print(n,"是素数。")
else:
        print(n,"是合数。")
```

# 9.问题：找出300以内的所有素数

```python
startNumber = 2
endNumber = 300
totalCount =0
print("在",startNumber,"~",endNumber,"中，符合条件的素数有：")
for n in range(startNumber,endNumber+1):
    for i in range(startNumber,n):
        if n%i == 0:
            break
    else:
            totalCount +=1
            print(n,end=" ")
print()
print("其中素数一共有",totalCount,"个。")
```

# 10.赌场有一种游戏，称为“幸运7”，游戏规则是你丢两个骰子，如果其点数之和为7你就赢4元，不是7你就输1元，请你分析一下，这样的规则是否公平呢？
##	a.用代码模拟10000次，给出获胜概率

```python
import random
WinCount = 0
TotalCount = 100000
for i in range(TotalCount):
    numFirst = random.randint(1,6)
    numSecond = random.randint(1,6)
    if numFirst + numSecond == 7:
        WinCount+=1
print("您的获胜概率是",WinCount/TotalCount)
```

##	b.假设刚开始有10元，当全部输掉为0元的时候游戏结束，用一段代码模拟参加游戏的过程。

```python
import random
TotalCount = 100000
Money = 10
MaxWin = Money
for i in range(TotalCount):
    if Money == 0:
        break
    else:
        numFirst = random.randint(1,6)
        numSecond = random.randint(1,6)
        if numFirst + numSecond == 7:
            Money += 4
            if MaxWin < Money:
                MaxWin = Money
        else:
            Money -= 1
        print("您的金额还有:",Money,"元。")
print("最大赢",MaxWin)
```

# 1.找出1000以内所有的水仙花数。水仙花数是指一个 3 位数，它的每个位上的数字的 3次幂之和等于它本身（例如：13 + 53+ 33 = 153）。

```python
startNumber = 100
endNumber = 1000
countTotal = 0
for i in range(startNumber,endNumber):
    firstBit = int(i/100)
    secondBit = int(i%100/10)
    thirdBit = int(i%10/1)
    if firstBit**3+secondBit**3+thirdBit**3 == i:
        print(i,end = " ")
        countTotal += 1
print("在",endNumber,"以内，共有水仙花数",countTotal,"个。")
```

# 2.问题：找出1000以内所有的完全数。完全数（Perfect number），是一些特殊的自然数。它所有的真因子（即除了自身以外的因子）的和（即因子函数），恰好等于它本身。

```python
startNumber = 1
endNumber = 1000
totalCount =0
print("在",startNumber,"~",endNumber,"中，符合条件的完全数有：")
for n in range(startNumber,endNumber+1):
    sumTotal = 0
    for i in range(startNumber,n):
        if n%i == 0:
            sumTotal += i
    if sumTotal == n:
        totalCount += 1
        print(n,end = " ")
print()
print("其中完全数一共有",totalCount,"个。")
```

# 3.问题：无穷级数4/1 - 4/3 + 4/5 - 4/7 + ……的和是圆周率π，请写一个程序计算出这一级数前n项的和。

```python
CalcPi = 0
endN = int(input("请输入您想要执行到的步数："))
for i in range(1,endN+1):
    CalcPi += (4*(-1)**(i+1)*(1/(2*i-1)))
print("经过系统计算，到第",endN,"步，π的值是",CalcPi,"。")
```

# 4.问题：无穷级数4/1 - 4/3 + 4/5 - 4/7 + ……的和是圆周率π，请写一个程序计算出这一级数前n项的和。但是要求改成：求出的圆周率π误差小于10-6的时候停止计算，输出求得的圆周率π值是多少？

```python
import math
CalcPi = 0
endN = 1
while abs(CalcPi - math.pi) >= 1e-6:
    CalcPi += (4*(-1)**(endN+1)*(1/(2*endN-1)))
    endN += 1
print("经过系统计算，到第",endN,"步，π的值是",CalcPi,"。误差小于10-6。")
```

# 5.问题：斐波那契数列（Fibonacci sequence），指的是这样一个数列：1、1、2、3、5、8、13、21、34、……。这个数列从第3项开始，每一项都等于前两项之和。现要求输出该数列的前n项，每行输出4个数字。

```python
endN = int(input("请输入您想要执行到的项数："))
FitonacciFirst = 1
FitonacciSecond = 1
FitonacciThird = 0
countTotal = 2
print("{:>30}{:>30}".format(FitonacciFirst,FitonacciSecond),end = " ")
for i in range (3,endN+1):
    FitonacciThird = FitonacciFirst+FitonacciSecond
    print("{:>30}".format(FitonacciThird),end = " " )
    countTotal += 1
    FitonacciFirst,FitonacciSecond = FitonacciSecond,FitonacciThird
    if countTotal % 4 == 0 :
        print()
```

```python
def fitonacci(countTotal):
    if countTotal == 1 or countTotal == 2:
        return 1
    else:
        return fitonacci(countTotal-1) + fitonacci(countTotal-2)
endN = int(input("请输入您想要执行到的项数："))
for i in range(1,endN+1):
    print(fitonacci(i))
```
