# 1.问题：由用户任意输入两个整数，求这两个数的和以及平均值是多少。

```python
a = eval(input("请输入第一个数："))
b = eval(input("请再输入第二个数："))
print("a=",a ,"b=",b)
sum = a+b
avg = sum/2
print("a+b的和是：",sum,"a+b的平均值是：",avg)
```

# 2.问题：需要找钱给用户，现在只有50元，5元和1元的人民币若干张。输入一个整数金额值，给出找钱的方案，假设人民币足够多，且优先使用面额大的钱币。

```python
money = eval(input("请输入您的金额（注意：整数！）："))
money_50 = money // 50
money_5 = (money % 50) // 5
money_1 = ((money % 50)%5) //1
print("您的金额为：",money,"应该找",money_50,"张50元面值的，应该找",money_5,"张5元面值的，应该找",money_1,"张1元面值的。")
```

# 3.问题：输入一个1-12的整数，输出对应的月份名称缩写。

```python
month_number = eval(input("请输入您需要转换的月份数字（注意：整数！）："))
# list_month = ["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"]
list_month = "JanFebMarAprMayJunJulAugSepOctNovDec"
month = list_month[(month_number-1)*3:(month_number-1)*3+3]
print("您的月份是：",month)
```

# 6.输入跑步成绩，跳绳成绩和跳远成绩，计算并输出体测总成绩。
# 总成绩为跑步成绩占40%，跳绳成绩占30%，跳远成绩占30%。

```python
Run = eval(input("请输入你的跑步成绩："))
JumpRope = eval(input("请输入你的跳绳成绩："))
LongJump = eval(input("请输入你的跳远成绩："))
TotalScore = Run*0.4+JumpRope*0.3+LongJump*0.3
InputText = "您跑步成绩是{}，您的跳绳成绩是{}，您的跳远成绩是{}。".format(Run,JumpRope,LongJump)
print(InputText)
print("由您的输入计算得到,您的体测成绩是",TotalScore,"。")
```
