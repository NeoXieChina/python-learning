````python
class Book():
    """Define a Book class"""
    def __init__(self, name:str,author:str,isbn:str,price:float,publisher:str):
        """Init the properties
        Args:
            name (string): [Book's name]
            author (string): [Book's author]
            isbn (string): [Book's isbn]
            price (float): [Book's price]
            publisher (str, optional): [description]. Defaults to "ABC Press".
        """
        self.name = name
        self.author = author
        self.isbn = isbn
        self.price = price
        self.publisher = publisher

    def info(self):
        print(f"The book's name:{self.name};\nThe book's author:{self.author};\nThe book's publisher:{self.publisher};\nThe book's isbn:{self.isbn};\nThe book's price:{self.price};")
help(Book)
````

## 3. 请定义一个TextBook类，从Book类继承；教材类是普通书籍类的一个子类拥有自己的属性：courseware(课件数，int)和exercise(练习题数: int)重载info()方法，增加courseware和exercise信息输出

```python
class TextBook(Book):  # 从Book类继承
    """TextBook(教材类)是普通书籍类的一个子类
       从Book类继承
    """
    def __init__(self, name, ...，courseware=18, ...):
	"""

	直接用类名调用父类方法在使用单继承的时候没问题，但是如果使用多继承，会涉及到查找顺序（MRO）、
	重复调用（钻石继承）等种种问题。
	"""
        super().__init__(name, author, isbn, price, publisher)

```



````python
class TextBook(Book):  # 从Book类继承
    """TextBook(教材类)是普通书籍类的一个子类
       从Book类继承
    """
    def __init__(self, name:str,author:str,isbn:str,price:float,publisher:str,courseware:int,exercise:int):
        """
    	super() 函数是用于调用父类(超类)的一个方法。super 是用来解决多重继承问题的，
    	直接用类名调用父类方法在使用单继承的时候没问题，但是如果使用多继承，会涉及到查找顺序（MRO）、
    	重复调用（钻石继承）等种种问题。
        """
        super().__init__(name, author, isbn, price, publisher)
        self.courseware = courseware # 教材会配套课件
        self.exercise = exercise  # 教材会配套习题

    def info(self):
        super().info()
        print(f"The textbook's courseware:{self.courseware};\nThe textbook's exercise:{self.exercise};")
````

````python
textbookNew =TextBook('Deep Learning with Python','Francois Chollet','9781617296864',59.5,'Manning Publications',12,200)
TextBook.info(textbookNew)
````


````python
class python_textbook(TextBook):
    def __init__ (self, name:str,author:str,isbn:str,price:float,publisher:str,courseware:int,exercise:int):
        super().__init__(name, author, isbn, price, publisher,courseware,exercise)
    # set_courseware()方法，用于设置课件数量
    def set_courseware(self, nums):
        self.courseware = nums  # 设置课件数量

    # get_ courseware()方法，用于获得课件数量
    def get_courseware(self):
        return self.courseware  # 获得课件数量
    @property
    def exercise(self):
        return self._exercise
    @exercise.setter
    def exercise(self, nums):
        if nums > 0:
            self._exercise = nums  # 若输入大于0，更新练习题数量
        else:
            self._exercise = 0   # 若输入小于等于0，设置为默认值
    def info(self):
        super().info()
python_textbookNewexercise =python_textbook('Deep Learning with Python','Francois Chollet','9781617296864',59.5,'Manning Publications',12,200)
# 练习题数量设置为666
python_textbook.exercise = 666
python_textbook.info(python_textbookNewexercise) 
# 实例化一个python_textbook对象，并调用info()方法，输出信息
````
