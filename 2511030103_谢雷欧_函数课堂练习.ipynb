# 1)找出2到100中所有的素数。

```python
def primeFinder(startNumber,endNumber):
    if startNumber > endNmuber:
        return
    else:
        totalCount =0
        primeList = []
        for n in range(startNumber,endNumber+1):
            for i in range(startNumber,n):
                if n%i == 0:
                    break
            else:
                totalCount +=1
                primeList.append(n)
        return [primeList,totalCount]
print("其中素数有：",primeFinder(2,100)[0],",一共",primeFinder(2,100)[1],"个。")
```

```python
def prime(Number):
    for i in range(2,Number):
        if Number%i ==0 :
            return False
    else:
        return True
for i in range(2,100+1):
    if prime(i) == True:
        print("{:^4}".format(i),end = " ")
```

# 2)找出2到100中所有的孪生素数。孪生素数就是指相差2的素数对，例如3和5，5和7，11和13等。

```python
def twinPrime(startNumber,endNmuber):
    if startNumber > endNmuber:
        return
    else:
        primeList = primeFinder(startNumber,endNmuber)[0]
        totalCount =0
        twinPrimeList = []
        for prime in range(len(primeList)):
            if (abs((primeList[prime] - primeList[prime-1])) == 2):
                twinPrimeList.append((primeList[prime-1],primeList[prime]))
                totalCount += 1
        return [twinPrimeList,totalCount]
print("其中孪生素数有：",twinPrime(2,100)[0],",一共",twinPrime(2,100)[1],"对。")
```

# ex4.1 基于期中成绩和期末成绩，按指定的权重计算总评成绩。(默认参数)
a. 期末占60%。期末占40%<br>
b. 期中和期末各占50%<br>
c. 期末占40%。期末占60%

```python
def calcScore(middleScore,finalScore,calcType='b'):
    if calcType == 'a':
        middleScorePercent= 0.6 
        finalScorePercent = 0.4
    elif calcType == 'b':
        middleScorePercent= 0.5
        finalScorePercent = 0.5
    elif calcType == 'c':
        middleScorePercent= 0.4
        finalScorePercent = 0.6
    else:
        return
    return middleScore*middleScorePercent+finalScore*finalScorePercent
middleScore = eval(input("请输入期中成绩："))
finalScore = eval(input("请输入期末成绩："))
print("a. 期末占60%。期末占40%")
print("b. 期中和期末各占50%")
print("c. 期末占40%。期末占60%")
calcType = input("请选择您想要按哪种方式计算（a,b,c):")
if not calcType:
    print("您的期中成绩是",middleScore,"，期末成绩是",finalScore,"，按照您指定的权重计算后得出的分数是",calcScore(middleScore,finalScore),"。")
else:
    print("您的期中成绩是",middleScore,"，期末成绩是",finalScore,"，按照您指定的权重计算后得出的分数是",calcScore(middleScore,finalScore,calcType),"。")
```

## 编写函数：返回输入的人名，并统计输入人名的个数

```python
def peopleCount(*peopleName):
    peopleNameList = []
    totalCount = 0
    for people in peopleName:
        peopleNameList.append(people)
        totalCount += 1
    return [peopleNameList,totalCount]
result = peopleCount("张三","李四","王五","李白","杜甫","王维")
print("您输入的人物有:",result[0],",一共",result[1],"人。")
```

# 编写一个函数来实现自动将利息添加到账户余额
hint：    money = money * (1 + rate)

```python
def addInterest(money,rate):
    money = money * (1 + rate )
    return money
amount = 1000
rate = 0.05
amount = addInterest(amount,rate)
print("amount=",amount)
```

# 列表的形参与实参
hint：amount = [1200,1400,800,650,1600]

```python
def addInterestList(money,rate):
    for i in range(len(money)):
        money[i] = money[i]*(1+rate)
amountList = [1200,1400,800,650,1600]
rate = 0.05
addInterestList(amountList,rate)
print("amount=",amountList)
```
