# 参考：计算三角形面积代码

```python
import math
a = float(input("请输入三角形的边长a："))

b = float(input("请输入三角形的边长b："))

c = float(input("请输入三角形的边长c："))

h = (a+b+c)/2

area = math.sqrt(h*(h-a)*(h-b)*(h-c))

print("三角形的面积为：{:.2f}".format(area))
```

# 3.问题：输入三角形三边，求三角形面积，对用户输入的数据做合法性检查。

```python
import math
a = float(input("请输入三角形的边长a："))
b = float(input("请输入三角形的边长b："))
c = float(input("请输入三角形的边长c："))

if a+b>c and a+c>b and b+c>a and a>0 and b>0 and c>0 and a-b<c and a-c<b and b-c<a:
    h = (a+b+c)/2
    area = math.sqrt(h*(h-a)*(h-b)*(h-c))
    print("三角形的面积为：{:.2f}".format(area))
else:
    print("您输入的数据有误，不是三角形！请重新输入！")
```

# 4.问题：根据用户的身高和体重，计算用户的BMI值，并给出相应的健康建议。
## BMI，即身体质量指数，是用体重（千克）除以身高（米）的平方得出的数字（BMI=体重(KG)÷身高2（M）），是目前国际上常用的衡量人体胖瘦程度以及是否健康的一个标准。我们先来看看成人的BMI数值：
### 成人的BMI数值
### 过轻：    低于18.5
### 正常：    18.5-23.9
### 过重：    24-27.9
### 肥胖：    28-32
### 过于肥胖：    32以上

```python
height = eval(input("请输入您的身高（米）："))
weight = eval(input("请输入您的体重（千克）："))
bml = weight/(height*height)
if bml<18.5:
    stand = "过轻"
elif bml<23.9:
    stand = "正常"
elif bml<27.9:
    stand = "过重"
elif bml<32:
    stand = "肥胖"
else:
    stand = "过于肥胖"
print("您的身高为"+str(height)+"米，您的体重为"+str(weight)+"千克，您的BMI为"+str(bml)+"，经过系统判断，您当前"+stand+"。")
```

# 5.问题：从键盘输入三位的正整数，输出其中的最大的一位数字是多少。例如：输入386，输出8；输入290，输出9。

```python
numberT= int(input("请输入一个三位的正整数："))
if numberT <=0:
    print("您输入的不是三位的正整数：")
else:
    first = int(numberT//100)
    second = int(numberT%100/10)
    third = int(numberT%10/1)
    if first>second and first>third and first==second==third:
        print("您输入的是{}，最大的一位数是{}。".format(numberT,first))
    elif second>first and second>third:
        print("您输入的是{}，最大的一位数是{}。".format(numberT,second))
    elif third>first and third>second:
        print("您输入的是{}，最大的一位数是{}。".format(numberT,third))
```

# 附加题
## A. 2输入小明百分制成绩，输出相应的等级：90分以上为‘A’，80分到89分为‘B’，70分到7分为‘C’，60分到69分为‘D’，60分以下为‘E’，如果分数大于100或者小于0，则输出“成绩有误”。

```python
score = eval(input("小明，请告诉我你的成绩："))
if score>100 or score<0:
    print("输入有误！")
else:
    if score>=90:
        grade = 'A'
    elif score>=80:
        grade = 'B'
    elif score>=70:
        grade = 'C'
    elif score>=60:
        grade = 'D'
    else:
        grade = 'E'
    print("你的等级是：", grade)
```
