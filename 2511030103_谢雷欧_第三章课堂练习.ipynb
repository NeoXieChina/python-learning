# 4.问题：根据用户的身高和体重，计算用户的BMI值，并给出相应的健康建议。
## BMI，即身体质量指数，是用体重（千克）除以身高（米）的平方得出的数字（BMI=体重(KG)÷身高2（M）），是目前国际上常用的衡量人体胖瘦程度以及是否健康的一个标准。我们先来看看成人的BMI数值：
### 成人的BMI数值
### 过轻：    低于18.5
### 正常：    18.5-23.9
### 过重：    24-27.9
### 肥胖：    28-32
### 过于肥胖：    32以上

```python
height = eval(input("请输入您的身高（米）："))
weight = eval(input("请输入您的体重（千克）："))
bml = weight/(height*height)
if bml<18.5:
    stand = "过轻"
elif bml<23.9:
    stand = "正常"
elif bml<27.9:
    stand = "过重"
elif bml<32:
    stand = "肥胖"
else:
    stand = "过于肥胖"
print("您的身高为"+str(height)+"米，您的体重为"+str(weight)+"千克，您的BMI为"+str(bml)+"，经过系统判断，您当前"+stand+"。")
```

# 1. 问题：统计英文句子中大写，小写和数字各有多少个？

```python
inputStr = str(input("请输入一串英文字符："))
countUpper=countLower=countNumber=0
for s in inputStr:
    if s.isupper():
        countUpper+=1
    elif s.islower():
        countLower+=1
    elif s.isdigit():
        countNumber+=1
print("在您输入的字符串",inputStr,"中，小写字符共",countLower,"个，大写字符共",countUpper,"个，数字字符共",countNumber,"个。")
```

# 2.问题：利用for循环求1至100中奇数和偶数的和分别是多少。

```python
sumOdd = sumEven = 0
startNumber = 1
endNumber = 100
for intNumber in range(startNumber,endNumber+1):
    if intNumber>0:
        if intNumber%2 == 0:
            sumEven+=intNumber
        else:
            sumOdd+=intNumber
print("在",startNumber,"-",endNumber,"中，奇数之和是",sumOdd,"，偶数之和是",sumEven,"。")
```

# 6.问题： “牛顿迭代法”求平方根。
## 1）假设需要求实数x的平方根，任取y为某个实数值
## 2）如果y×y=x，计算结束，y就是x的平方根
## 3）否则令z=(y+x/y)/2
## 4）令y的新值等于z，转回步骤1）
## y=(y+x/y)/2

```python
import math
x = float(input("请输入一个数赋值给x:"))
y = 1.0
while abs(y*y - x)>10e-8:
        y = (y+x/y)/2
print("根据牛顿迭代法计算，",x,"的平方根是",y,"。")
print("使用函数计算，",x,"的平方根是",math.sqrt(x),"。")
```

# 7.问题：判断一个正整数n（n>=2）是否为素数。素数又称质数。一个大于1的自然数，除了1和它自身外，不能被其它整数整除的数叫做素数；否则称为合数。

```python
n = int(input("请输入一个正整数："))
startNumber = 2
for i in range(startNumber,n):
    if n%i == 0:
        break
if i == n-1:
        print(n,"是素数。")
else:
        print(n,"是合数。")
```
