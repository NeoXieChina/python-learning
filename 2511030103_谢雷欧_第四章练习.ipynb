# d.反向排序列表，输出列表，并说明此时列表是否还是原列表

```python
listSecond = sorted(listFirst,reverse = True)
print("修改后:",listSecond)
if listFirst == listSecond:
    print("排序后的列表是源列表。")
else:
    print("排序后的列表不是源列表。")
```

# 3. 习题3 给出某市刮刮乐2006-2015十年间的中奖率，(2006,57),(2007,56),(2008,57),(2009,62),(2010,69),(2011,72),(2012,75),(2013,76),(2014,74.3),(2015,74)
# 请用代码统计这十年平均中奖率，以及并输出中奖率最高的年份及其中奖率（用列表实现）

```python
listWin = [[2006,57],[2007,56],[2008,57],[2009,62],[2010,69],[2011,72],[2012,75],[2013,76],[2014,74.3],[2015,74]]
maxWinYear = listWin[0][0]
maxWinProbability = listWin[0][1]
for i in range(len(listWin)):
    print(listWin[i][0],"年的中奖率是：",listWin[i][1])
    if maxWinProbability <  listWin[i][1]:
        maxWinProbability = listWin[i][1]
        maxWinYear = listWin[i][0]
print("中奖率最高是",maxWinProbability,"是在",maxWinYear,"创造的。")
```

# 4. 习题4 生成的20个1000~5000的随机数，并汇总在一个列表中，并输出在这个列表中找到不能被10以内素数整除的数，将这些数汇总在另一个列表中，并输出

```python
import random
listRangeItem = [random.randint(1000,5000)]
for i in range(19):
    listRangeItem.append(random.randint(1000,5000))
print("生成的列表为：",listRangeItem)
listPrime = [2]
for n in range(3,10):
    is_Prime = True
    for i in range(2,n):
        if n%i == 0:
            is_Prime = False
            break
    if is_Prime == True:
        listPrime.append(n)
print(listPrime)
listNotDivByPrime = []
for n in range(len(listRangeItem)):
    divable = False
    for i in range(len(listPrime)):
        if 0 == listRangeItem[n]%listPrime[i]:
            divable = True
            break
    if divable == False:
        listNotDivByPrime.append(listRangeItem[n])
print(listNotDivByPrime)
```
