# 集合运算

```python
setIEEE = {'Python','C++','C','Java','C#'}
setTIOBE = {'Java','C','Python','C++','VB.NET'}
print("IEEE2018排行榜前五的编程语言有：")
print(setIEEE)
print("TIOBE2018排行榜前五的编程语言有：")
print(setTIOBE)
print("前五名上榜的所有语言有：")
print(setIEEE | setTIOBE)
print("在两个榜单同时进前五的语言有")
print(setIEEE & setTIOBE)
print("只在IEEE榜进前五的语言有")
print(setIEEE - setTIOBE)
print("只在一个榜单进前五的语言")
print(setIEEE ^ setTIOBE)
```

# （3）本学期学校共开设了3门选修课，一个班有25位学生，选修的情况如下：
## 选修1号课程的同学有：李雷、张玉、王晓刚、陈红静、方向、司马清；
## 选修2号课程的同学有：施然、李芳芳、刘潇、方向、孙一航、黄煌；
## 选修3号课程的同学有：陈红静、方向、刘培良、张玉、施小冉、司马清。
## 请编写程序解决以下问题：
## ①这个班有多少位学生没有选课?

```python
classChosed = {'class1': ["李雷","张玉","王晓刚","陈红静","方向","司马清"],'class2':["施然","李芳芳","刘潇","方向","孙一航","黄煌"],'class3':["陈红静","方向","刘培良","张玉","施小冉","司马清"]}
studentChosed = {}
for classId in classChosed:
    for student in classChosed[classId]:
        studentChosed[student] = studentChosed.get(student,0) + 1
print("这个班有",(25-len(studentChosed)),"个学生没有选上课。")
```

## ②有多少位学生同时选修了2门课?

```python
studentCount = 0
for student in studentChosed:
    if studentChosed[student] == 2:
      studentCount += 1
print("有",studentCount,"位学生同时选修了2门课。")
```

## ③有多少位学生同时选修了3门课？

```python
studentCount = 0
for student in studentChosed:
    if studentChosed[student] == 3:
      studentCount += 1
print("有",studentCount,"位学生同时选修了3门课。")
studentCount = len(set(classChosed['class1']) & set(classChosed['class2']) & set(classChosed['class3']))
print(studentCount)
```

## ④有多少位学生只选修了1门课？

```python
studentCount = 0
for student in studentChosed:
    if studentChosed[student] == 1:
      studentCount += 1
print("有",studentCount,"位学生同时选修了1门课。")
```
